```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as mlt
import random
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Dense, Flatten
```

```python
X_train = np.loadtxt("C:\\GEN1_AI\\Dataset\\Vision\\Image_Classification_CNN_Keras_Dataset\\input.csv", delimiter = ',')
Y_train = np.loadtxt("C:\\GEN1_AI\\Dataset\\Vision\\Image_Classification_CNN_Keras_Dataset\\labels.csv", delimiter = ',')

X_test = np.loadtxt("C:\\GEN1_AI\\Dataset\\Vision\\Image_Classification_CNN_Keras_Dataset\\input_test.csv", delimiter = ',')
Y_test = np.loadtxt("C:\\GEN1_AI\\Dataset\\Vision\\Image_Classification_CNN_Keras_Dataset\\labels_test.csv", delimiter = ',')
```

```python
X_train = X_train.reshape(len(X_train), 100, 100, 3)
Y_train = Y_train.reshape(len(Y_train), 1)

X_test = X_test.reshape(len(X_test), 100, 100, 3)
Y_test = Y_test.reshape(len(Y_test), 1)

X_train = X_train/255.0
```

```python
print("X_train",X_train.shape)
print("Y_train",Y_train.shape)
print("\n")
print("X_test",X_test.shape)
print("Y_test",Y_test.shape)
```

```python
X_train[1,:]
```

```python
idx = random.randint(0, len(X_train))
mlt.imshow(X_train[idx, :])
mlt.show()
```

```python
model = Sequential([
    Conv2D(32, (3,3), activation = 'relu'),
    MaxPooling2D((2,2)),

    Conv2D(32, (3,3), activation = 'relu'),
    MaxPooling2D((2,2)),

    Flatten(),
    Dense(64, activation = 'relu'),
    Dense(1, activation = 'sigmoid')
])
```

```python
model = Sequential()

model.add(Conv2D(32, (3,3), activation = 'relu', input_shape = (100, 100, 3)))
model.add(MaxPooling2D((2,2)))

model.add(Conv2D(32, (3,3), activation = 'relu'))
model.add(MaxPooling2D((2,2)))

model.add(Flatten())
model.add(Dense(64, activation = 'relu'))
model.add(Dense(1, activation = 'sigmoid'))
```

```python
# opt = keras.optimizers.SGD(learning_rate = 0.001)
model.compile(loss='binary_crossentropy', optimizer='adam', metrics = ['accuracy'])
```

```python
model.fit(X_train, Y_train, epochs = 50, batch_size = 64 )
```

```python
model.evaluate(X_test, Y_test)
```
