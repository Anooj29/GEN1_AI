<h4>We will be using <b>train/test split. Let's import <b>train_test_split</b> from <b>sklearn.model_selection</b>.

```python
from sklearn.model_selection import train_test_split
```

Now <b> train_test_split </b> will return 4 different parameters. We will name them:<br>
x_trains, x_test, y_train, y_tests <br> <br>

When we split your data into x and y, we are essentially separating the features (inputs) from the target (output) variable. 

<b>x:</b> This represents the input data that the model will learn from. -- <b><u>INDEPENDENT VARIABLE</u></b><br>
<b>y:</b> This represents the output variable that you want to predict. -- <b><u>DEPENDENT VARIABLE</u></b>

Here, test_size=0.2 which means 20% of the data is for testing, i.e., 20% of the data will be reserved for testing, while 80% will be used for training.

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=51)
```

<h4>As, we can observe out of 569 rows <b>114 rows (20%)</b> are reserved for test split and <b>455 rows (80%)</b> are for train.

```python
x_test.shape
```

```python
x_train.shape
```

<h3>StandardScaler</h3><br>
<h4>Use for standardizing features by removing the <b>mean and scaling to unit variance</b>. This is an important preprocessing step in many machine learning algorithms, as it helps to ensure that the model performs optimally.<br></h4>
StandarScaler is use for following;<br>
- Standardization<br>
- Handling Different Scales<br>
- Fitting and Transforming

```python
from sklearn.preprocessing import StandardScaler 
sc = StandardScaler()
```

<h4>Use the fit method to compute the <b>mean and standard deviation</b> for each feature in our dataset. This step is essential because it allows the scaler to learn the parameters needed for standardization.

```python
sc.fit(x_train)
```

<h4>'transform' method will apply the standardization process to our dataset.
<br></br>
 also we can use  <u>fit_transform</u> method to combine the fitting and transforming steps into one.

```python
x_train = sc.transform(x_train)
x_test = sc.transform(x_test)
```

```python
x_test
```

```python
x_train
```

```python
from sklearn.linear_model import LogisticRegression
lg = LogisticRegression()
```

```python
lg.fit(x_train, y_train)
y_pred = lg.predict(x_test)
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test, y_pred)
```
